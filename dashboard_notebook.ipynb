__Covid19 Dashboard__

In this notebook I'll build an interactive dashboard for COVID19. Firstly, let's prepare the data before visualizing

```python
# loading confirmed cases, deaths and recoveries from GitHub repository
import pandas as pd
confirmed_cases_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
deaths_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recoveries_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

Well, cases are not convinently distributed (they are in wide format). It will be transformed into a long format in the next section

```python
confirmed_cases_data = pd.melt(confirmed_cases_data, id_vars=confirmed_cases_data.columns[:4], value_vars=confirmed_cases_data.columns[4:], var_name='date', value_name='confirmed')
deaths_data = pd.melt(deaths_data, id_vars=deaths_data.columns[:4], value_vars=deaths_data.columns[4:], var_name='date', value_name='deaths')
recoveries_data = pd.melt(recoveries_data, id_vars=recoveries_data.columns[:4], value_vars=recoveries_data.columns[4:], var_name='date', value_name='recoveries')
```

The problem is that in Deaths/Confirmed cases datasets Canada's data is represented with provinces while in Recoveries provinces are not mentioned. To continue working on the data, this possible conflict must be resolved

```python
# summarizing numbers in Canada confirmed cases and deaths
confirmed_cases_canada = confirmed_cases_data[confirmed_cases_data['Country/Region']=='Canada'].groupby('date').sum()[['confirmed']]
deaths_canada = deaths_data[deaths_data['Country/Region']=='Canada'].groupby('date').sum()[['deaths']]

# extracting Canada recoveries (columns)
recoveries_canada = recoveries_data[recoveries_data['Country/Region']=='Canada']
template_updated_canada = recoveries_canada[recoveries_canada.columns[:-1]].reset_index(drop=True)

# joining deaths and confirmed cases with teh columns
confirmed_cases_canada = template_updated_canada.merge(confirmed_cases_canada, how='inner', left_on='date', right_index=True)
deaths_canada = template_updated_canada.merge(deaths_canada, how='inner', left_on='date', right_index=True)

# altering the original deaths and confirmed cases to resolve the problem
confirmed_cases_data = confirmed_cases_data[confirmed_cases_data['Country/Region']!='Canada'].append(confirmed_cases_canada)
deaths_data = deaths_data[deaths_data['Country/Region']!='Canada'].append(deaths_canada)
```

The problem was resolved by summarizing deaths and confirmed cases for the whole country by date and merging with recoveries values

Then all the three tables will be cobined into one table

```python
combined_data = confirmed_cases_data.merge(deaths_data, how='inner', on=['Country/Region', 'Province/State', 'date', 'Lat', 'Long']).merge(recoveries_data, how='inner', on=['Country/Region', 'Province/State', 'date', 'Lat', 'Long'])
combined_data = combined_data.drop(['Lat', 'Long'], axis=1)
combined_data.sample(15)
```

Time to build the dash board

To build an interactive map, first, we need to convert countries to ISO codes (alpha-3)

```python
import pycountry
contries_lst = [i.name for i in  pycountry.countries]
for country in combined_data['Country/Region'].unique():
    if country not in contries_lst:
        print(country)
```

```python
# building a dictionary to convert countries and omit non-country values
omit_lst = ['Diamond Princess', 'Kosovo', 'MS Zaandam', 'Summer Olympics 2020', 'Winter Olympics 2022']
convert_dict = {
    'Congo (Brazzaville)': 'Congo',
    'Congo (Kinshasa)': 'Congo',
    "Cote d'Ivoire": "Côte d'Ivoire",
    'Bolivia':'Bolivia, Plurinational State of',
    'Holy See':'Holy See (Vatican City State)',
    'Iran':'Iran, Islamic Republic of',
    'Korea, North':"Korea, Democratic People's Republic of",
    'Korea, South': 'Korea, Republic of',
    'Laos':"Lao People's Democratic Republic",
    'Micronesia':'Micronesia, Federated States of',
    'Moldova':'Moldova, Republic of',
    'Moldova':'Moldova, Republic of',
    'Russia':'Russian Federation',
    'Taiwan*':'Taiwan, Province of China',
    'Tanzania':'Tanzania, United Republic of',
    'US': 'United States',
    'West Bank and Gaza': 'Israel',
    'Burma': 'Myanmar',
    'Venezuela':'Venezuela, Bolivarian Republic of',
    'Vietnam':'Viet Nam'
}
# omitting and converting
combined_data['Country/Region']=combined_data['Country/Region'].replace(convert_dict)
for el in omit_lst:
    combined_data = combined_data.drop(combined_data[combined_data['Country/Region'] == el].index)
```

```python
# convert test
import pycountry_convert 
for i in combined_data['Country/Region'].unique():
    print(pycountry_convert.country_name_to_country_alpha3(i))
```

```python
#creating ISO-codes column 
combined_data['ISO-codes'] = combined_data.apply(lambda x: pycountry_convert.country_name_to_country_alpha3(x['Country/Region']), axis=1)
combined_data
```

```python
#total confirmed cases, deaths, recoveries for depicting on a map
map_data=combined_data.groupby('ISO-codes').sum()[['confirmed', 'deaths', 'recoveries']]
map_data['ISO-codes']=map_data.index
map_data.index=range(1, len(map_data)+1)
map_data['Country']=map_data.apply(lambda x: pycountry_convert.country_alpha2_to_country_name(pycountry_convert.country_alpha3_to_country_alpha2(x['ISO-codes'])), axis=1)
map_data
```

```python
# importing geojson data
import geopandas as gpd
geo_data = gpd.read_file('https://datahub.io/core/geo-countries/r/countries.geojson')
geo_data=geo_data.drop(['ADMIN'], axis=1)
geo_data=geo_data.rename(columns={'ISO_A3':'ISO-codes'})
geo_data.head()

# merging geojson with the data for our map
map_w_geo = geo_data.merge(map_data, how='inner', on='ISO-codes')
```

```python
# creating a map with layers
import folium
import branca.colormap as cm

# creating a map object and feature aka layers
full_covid_map = folium.Map(location=[13, 16], zoom_start=2, overlay=False)
confirmed_feature = folium.FeatureGroup(name='Confirmed cases', overlay=False).add_to(full_covid_map)
deaths_feature = folium.FeatureGroup(name='Deaths', overlay=False).add_to(full_covid_map)
recoveries_feature = folium.FeatureGroup(name='Recoveries', overlay=False).add_to(full_covid_map)

# building a confirmed cases layer
confirmed_layer = folium.Choropleth(geo_data=r'https://datahub.io/core/geo-countries/r/countries.geojson', 
                                    data=map_w_geo, 
                                    name='confirmed_cases',
                                    columns=['ISO-codes', 'confirmed'],
                                    key_on='feature.properties.ISO_A3',
                                    fill_color="YlGnBu",
                                    nan_fill_color="White",
                                    fill_opacity=0.7,
                                    line_opacity=0.2,
                                    legend_name="Confirmed cases (people)",
                                    highlight=True,
                                    overlay=False,
                                    line_color='black' ).geojson.add_to(confirmed_feature)
folium.features.GeoJson(
data=map_w_geo,
name='Total Confirmed Cases',
smooth_factor=2,
style_function=lambda x: {'color':'black','fillColor':'transparent','weight':0.5},
tooltip=folium.features.GeoJsonTooltip(
    fields=['Country',
            'confirmed'],
    aliases=["Country Name:",
            "Total confirmed cases:"], 
    localize=True,
    sticky=False,
    labels=True,
    style="""
        background-color: #F0EFEF;
        border: 2px solid black;
        border-radius: 3px;
        box-shadow: 3px;
    """,
    max_width=800,),
        highlight_function=lambda x: {'weight':3,'fillColor':'grey'},
    ).add_to(confirmed_layer)

# building deaths case layer
deaths_layer = folium.Choropleth(geo_data=r'https://datahub.io/core/geo-countries/r/countries.geojson', 
                                data=map_w_geo, 
                                name='deaths_cases',
                                columns=['ISO-codes', 'deaths'],
                                key_on='feature.properties.ISO_A3',
                                fill_color="YlOrRd",
                                fill_opacity=0.7,
                                nan_fill_color="White",
                                line_opacity=0.2,
                                legend_name="Deaths (people)",
                                highlight=True,
                                overlay=False,
                                line_color='black' ).geojson.add_to(deaths_feature)
folium.features.GeoJson(
data=map_w_geo,
name='Total Deaths',
smooth_factor=2,
style_function=lambda x: {'color':'black','fillColor':'transparent','weight':0.5},
tooltip=folium.features.GeoJsonTooltip(
    fields=['Country',
            'deaths'],
    aliases=["Country Name:",
            "Total deaths:"], 
    localize=True,
    sticky=False,
    labels=True,
    style="""
        background-color: #F0EFEF;
        border: 2px solid black;
        border-radius: 3px;
        box-shadow: 3px;
    """,
    max_width=800,),
        highlight_function=lambda x: {'weight':3,'fillColor':'grey'},
    ).add_to(deaths_layer)

# building recoveries layer
recoveries_layer = folium.Choropleth(geo_data=r'https://datahub.io/core/geo-countries/r/countries.geojson', 
                                    data=map_w_geo, 
                                    name='recoveries',
                                    columns=['ISO-codes', 'recoveries'],
                                    key_on='feature.properties.ISO_A3',
                                    fill_color='PuBuGn',
                                    nan_fill_color="White",
                                    fill_opacity=0.7,
                                    line_opacity=0.2,
                                    legend_name="Recoveries (people)",
                                    highlight=True,
                                    overlay=False,
                                    line_color='black').geojson.add_to(recoveries_feature)

folium.features.GeoJson(
data=map_w_geo,
name='TotalRecoveries',
smooth_factor=2,
style_function=lambda x: {'color':'black','fillColor':'transparent','weight':0.5},
tooltip=folium.features.GeoJsonTooltip(
    fields=['Country',
            'recoveries'],
    aliases=["Country Name:",
            "Total recoveries:"], 
    localize=True,
    sticky=False,
    labels=True,
    style="""
        background-color: #F0EFEF;
        border: 2px solid black;
        border-radius: 3px;
        box-shadow: 3px;
    """,
    max_width=800,),
        highlight_function=lambda x: {'weight':3,'fillColor':'grey'},
    ).add_to(recoveries_layer)  
folium.LayerControl(collapsed=False).add_to(full_covid_map)

# adding legends
# the way the numbers were calculated: map_w_geo['confirmed'].max() and billions were taken from it
colormap_confirmed = cm.linear.YlGnBu_08.scale(0, 40).to_step(10)
colormap_confirmed.caption = "Confirmed cases (billions)"
full_covid_map.add_child(colormap_confirmed)

# the way the numbers were calculated: map_w_geo['deaths'].max() and billions were taken from it
colormap_deaths= cm.linear.YlOrRd_05.scale(0, 500).to_step(100)
colormap_deaths.caption = "Deaths (millons)"
full_covid_map.add_child(colormap_deaths)

# the way the numbers were calculated: map_w_geo['recoveries'].max() and billions were taken from it
colormap_recoveries = cm.linear.PuBuGn_06.scale(0, 50).to_step(10)
colormap_recoveries.caption = "Recoveries (billions)"
full_covid_map.add_child(colormap_recoveries)


# saving to html file
full_covid_map.save('full_covid_map.html')
```
